```python
print(data1['Region of residence'].unique())
print(data1['Rating of Online Class experience'].unique())
print(data1['Medium for online class'].unique())
print(data1['Prefered social media platform'].unique())
print(data1['Change in your weight'].unique())
print(data1['Health issue during lockdown'].unique())
print(data1['Time utilized'].unique())
print(data1['Do you find yourself more connected with your family, close friends , relatives  ?'].unique())
```

```python
data1.nunique()
```

```python
data['Time spent on TV'].replace('No tv', 0, inplace=True)
data['Time spent on TV'].replace('n', 0, inplace=True)
data['Time spent on TV'].replace('N', 0, inplace=True)
data['Time spent on TV'].replace(' ', 0, inplace=True)
data['Time spent on TV'] = data['Time spent on TV'].astype('float')
data['Prefered social media platform'] = data['Prefered social media platform'].replace('None ', 'None')
data['Prefered social media platform'] = data['Prefered social media platform'].replace('Whatsapp', 'WhatsApp')
data.rename(columns = {"Health issue during lockdown":"Mental Health issue during lockdown"}, inplace = True)
```

```python
plt.figure(figsize=(20,10))
sns.displot(data1['Age of Subject'])
plt.title('Distribution of Age Subject')
plt.show()
```

```python
medium_val = data['Medium for online class'].value_counts().to_dict()
medium_val
```

```python
plt.figure(figsize=(12,8))
sns.set(style='ticks')
sns.countplot(y='Medium for online class', data=data1, 
              order=data1['Medium for online class'].value_counts().index[:5], palette='Set2')
plt.xlabel("Number of Students", weight='bold')
plt.ylabel("Medium of online class", weight='bold')
plt.show()
```

```python
dict_ = data['Prefered social media platform'].value_counts().to_dict()
plt.figure(figsize=(14,14))
plt.pie(x=dict_.values(),autopct='%1.1f%%', startangle=90,radius=0.75)
plt.legend(labels=dict_.keys(), loc='upper right',shadow=True, facecolor='lightyellow')
plt.show()
```

```python
# How they utilised their time?
plt.figure(figsize=(14,10))
sns.set(style='whitegrid')
sns.boxplot(data=data[['Time spent on Online Class','Time spent on self study','Time spent on fitness',
                       'Time spent on sleep','Time spent on social media', 'Time spent on TV']], orient='h', palette='Set3')
plt.show()
```

```python
sns.set(style='darkgrid')
fig, ax = plt.subplots(2,2, figsize=(16,18))
sns.violinplot(x='Mental Health issue during lockdown', y='Time spent on Online Class', data=data, ax=ax[0,0])
sns.violinplot(x='Mental Health issue during lockdown', y='Time spent on self study', data=data, ax=ax[0,1])
sns.violinplot(x='Mental Health issue during lockdown', y='Time spent on fitness', data=data, ax=ax[1,0])
sns.violinplot(x='Mental Health issue during lockdown', y='Time spent on sleep', data=data, ax=ax[1,1])
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.set(style='white')
sns.violinplot(y='Mental Health issue during lockdown', x='Number of meals per day', data=data)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(style='ticks')
sns.countplot(y='Stress busters', data=data, order=data['Stress busters'].value_counts().index[:15], palette='Set3')
plt.xlabel("Number of Students", weight='bold')
plt.ylabel("Stress buster activity", weight='bold')
plt.show()
```
